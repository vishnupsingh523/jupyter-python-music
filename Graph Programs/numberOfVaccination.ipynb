```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[2,4,6])
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x = [16,17,18,19,20,21,22]
y = [20000,23000,40000,57234,123456,105400,252089]
plt.plot(x,y,'bo',label='vaccinated people', linewidth=5)
plt.title('Vaccination details in India')
plt.xlabel('Days')
plt.ylabel('Number of people Vaccinated')
plt.legend()
plt.grid(True, color='black')
plt.show()
```
